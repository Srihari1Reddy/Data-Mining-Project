> ### Importing Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

> ### Reading csv file into dataframe

```python
df=pd.read_csv('../data/crdHT.csv')
```

> ### Total No of Cases in Each State in India

```python
sum_crime=[]
label=['ANDHRA PRADESH','ARUNACHAL PRADESH','ASSAM','BIHAR','CHHATTISGARH','GOA','GUJARAT','HARYANA','HIMACHAL PRADESH','JAMMU & KASHMIR','JHARKHAND','KARNATAKA','KERALA','MADHYA PRADESH','MAHARASHTRA','MANIPUR','MEGHALAYA','MIZORAM','NAGALAND','ODISHA','PUNJAB','RAJASTHAN','SIKKIM','TAMIL NADU','TRIPURA','UTTAR PRADESH','UTTARAKHAND','WEST BENGAL','A & N ISLANDS','CHANDIGARH','D & N HAVELI','DAMAN & DIU','PUDUCHERRY','LAKSHADWEEP','DELHI']
for i in label:
    state=df[df['STATE/UT']==i]
    total=sum(state['Cases reported during the year'])
    sum_crime.append(total)

plt.figure(figsize=(20,10))
index = np.arange(len(label))
plt.bar(index, sum_crime)
plt.xlabel('States', fontsize=15)
plt.ylabel('No of Cases', fontsize=15)
plt.xticks(index, label, fontsize=10, rotation=90)
plt.title('Total No of Cases in Each State')
plt.savefig('../plots/No_of_Cases_in_Each_State.png', dpi=300, bbox_inches='tight')
plt.show()
```

> ### Average No of Cases in Each State for Every Crime

```python
label=['IMMORAL TRAFFIC (PREVENTION) ACT','BUYING OF GIRLS FOR PROSTITUTION (SECTION 373 IPC)','SELLING OF GIRLS FOR PROSTITUTION (SECTION 372 IPC)','PROCURATION OF MINOR GIRLS (SECTION 366-A IPC)','IMPORTATION OF GIRLS FROM FOREIGN COUNTRY(SECTION 366-B IPC)','HUMAN TRAFFICKING']
avg_crime=[]
for i in label:
    crm_provider=df[df['Crime Head']==i]
    avg=sum(crm_provider['Cases reported during the year'])/len(crm_provider['Cases reported during the year'])
    avg_crime.append(avg)
    
plt.figure(figsize=(16,8))
index=np.arange(len(label))
plt.bar(index, avg_crime)
plt.xlabel('Crime Head',fontsize=10)
plt.ylabel('Average Number of Cases per State',fontsize=10)
plt.xticks(index, label, fontsize=10, rotation=45)
plt.title('Average')
plt.savefig('../plots/Average_No_of_Cases_per_State.png',dpi=300, bbox_inches='tight')
plt.show()
```

> ### No of Cases reported this Year vs No of Cases Pending Investigation from Last Year

```python
sum_crime=[]
sum_crime_pnd=[]
label=['ANDHRA PRADESH','ARUNACHAL PRADESH','ASSAM','BIHAR','CHHATTISGARH','GOA','GUJARAT','HARYANA','HIMACHAL PRADESH','JAMMU & KASHMIR','JHARKHAND','KARNATAKA','KERALA','MADHYA PRADESH','MAHARASHTRA','MANIPUR','MEGHALAYA','MIZORAM','NAGALAND','ODISHA','PUNJAB','RAJASTHAN','SIKKIM','TAMIL NADU','TRIPURA','UTTAR PRADESH','UTTARAKHAND','WEST BENGAL','A & N ISLANDS','CHANDIGARH','D & N HAVELI','DAMAN & DIU','PUDUCHERRY','LAKSHADWEEP','DELHI']
for i in label:
    state=df[df['STATE/UT']==i]
    total=sum(state['Cases reported during the year'])
    totalpnd=sum(state['Cases Pending Investigation from Previous Year'])
    sum_crime.append(total)
    sum_crime_pnd.append(totalpnd)

plt.figure(figsize=(25,10))
index = np.arange(len(label))
bar_width=0.50
plt.bar(index, sum_crime, bar_width,label = 'Cases reported in this Year')
plt.bar(index + bar_width, sum_crime_pnd, bar_width,label = 'Cases Pending Investigation From Last Year')
plt.xlabel('States', fontsize=15)
plt.ylabel('No of Cases', fontsize=15)
plt.xticks(index + [bar_width/2], label, fontsize=12, rotation=90)
plt.title('No of Cases reported this Year vs No of Cases Pending Investigation from Last Year')
plt.savefig('../plots/Comparisson_of_Reported_Cases_from_this_year_vs_last_year', dpi=300, bbox_inches='tight')
plt.legend()
plt.show()
```

 ### No of Cases Pending Trial At the End of this Year vs Cases Pending Trial from Previous Year

```python
sum_crime=[]
sum_crime_pnd=[]
label=['ANDHRA PRADESH','ARUNACHAL PRADESH','ASSAM','BIHAR','CHHATTISGARH','GOA','GUJARAT','HARYANA','HIMACHAL PRADESH','JAMMU & KASHMIR','JHARKHAND','KARNATAKA','KERALA','MADHYA PRADESH','MAHARASHTRA','MANIPUR','MEGHALAYA','MIZORAM','NAGALAND','ODISHA','PUNJAB','RAJASTHAN','SIKKIM','TAMIL NADU','TRIPURA','UTTAR PRADESH','UTTARAKHAND','WEST BENGAL','A & N ISLANDS','CHANDIGARH','D & N HAVELI','DAMAN & DIU','PUDUCHERRY','LAKSHADWEEP','DELHI']
for i in label:
    state=df[df['STATE/UT']==i]
    total=sum(state['Cases sent for trial during the year'])
    totalpnd=sum(state['Cases pending trial from the previous year'])
    sum_crime.append(total)
    sum_crime_pnd.append(totalpnd)

plt.figure(figsize=(25,10))
index = np.arange(len(label))
bar_width=0.50
plt.bar(index, sum_crime, bar_width,label = 'Cases for Trial in this year')
plt.bar(index + bar_width, sum_crime_pnd, bar_width,label = 'Cases Pending Trial From Last Year')
plt.xlabel('States', fontsize=15)
plt.ylabel('No of Cases in Tril', fontsize=15)
plt.xticks(index + [bar_width/2], label, fontsize=12, rotation=90)
plt.title('No of Cases at Trail in this Year vs Last Year')
plt.savefig('../plots/Comparisson_of_Trial_Cases_from_this_year_vs_last_year', dpi=300, bbox_inches='tight')
plt.legend()
plt.show()
```

### Percentage of Cases Convicted vs Cases Acquited/Discharged

```python
label=['IMMORAL TRAFFIC (PREVENTION) ACT','BUYING OF GIRLS FOR PROSTITUTION (SECTION 373 IPC)','SELLING OF GIRLS FOR PROSTITUTION (SECTION 372 IPC)','PROCURATION OF MINOR GIRLS (SECTION 366-A IPC)','IMPORTATION OF GIRLS FROM FOREIGN COUNTRY(SECTION 366-B IPC)','HUMAN TRAFFICKING']
avg_crime=[]
for i in label:
    crm_provider=df[df['Crime Head']==i]
    total=sum(crm_provider['Cases convicted'])
    totalaqt=sum(crm_provider['Cases acquitted or discharged'])
    sizes=[total,totalaqt]
    labels=['Cases convicted','Cases acquitted or discharged']
    avg_crime.append(label)
    explode =[0.1,0]
    plt.figure(figsize=(7,7))
    plt.xlabel(i)
    plt.title("Pie Plot Between Percentage of Cases Convicted vs Cases acquited/Discharged")
    plt.pie(sizes, labels=labels,explode=explode, autopct='%1.2f%%', shadow=True)
    plt.savefig('../plots/PiePlot_Percentage_Cases_Convicted_Acquited_or_Disharged.png',dpi=300, bbox_inches='tight')
    plt.show()
```
